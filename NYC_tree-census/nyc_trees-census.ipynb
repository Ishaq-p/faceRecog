## first of all lets fill NaNs for stumps and dead trees with 'not applicable'

```python
data12[data12.status!='Alive'].info()
```

```python
data12[data12.status!='Alive'][['health','spc_latin','steward','sidewalk','problems']].info()
```

```python
mask5 = (data12['status']=='Stump') | (data12['status']=='Dead')
```

```python
data12.loc[mask5, ['health','spc_latin','steward','sidewalk','problems']] = data12.loc[mask5, ['health','spc_latin','steward','sidewalk','problems']].fillna('notApplicable')
```

```python
data12[mask5]
```

```python
data12.isna().sum()
```

## here we're missing the 'health' value for a an 'Alive' tree
## so we're gonna just fill it with the most common one

```python
data12[data12.health.isna()]
```

```python
data12.health.value_counts()
```

```python
data12.loc[data12.health.isna()] = data12[data12.health.isna()].fillna('Good')
```

```python
data12[data12.health.isna()]
```

## here, i even saw it in a tutorial to fill them with 'no observartion', but it doesn't seem to be right thing to do and neither can we fill them with most common value, furthermore these are just 5 rows so i will just delete them

```python
data12[data12.spc_latin.isna()]
```

```python
data12.spc_latin.value_counts()
```

```python
data12.drop([356613, 427541, 431417, 608632, 656960], axis=0, inplace=True)
```

```python
data12[data12.spc_latin.isna()]
```

## here there's only one row which is missing the 'sidewalk' damage value, so we are about to fill it with the most common one

```python
data12[data12.sidewalk.isna()]
```

```python
data12.sidewalk.value_counts()
```

```python
data12.loc[data12.sidewalk.isna()] = data12[data12.sidewalk.isna()].fillna('NoDamage')
```

```python
data12[data12.sidewalk.isna()]
```

```python
data12.isna().sum()
```

## here for missing 'problems' values we will fill them with the most common ones

```python
data12[data12.problems.isna()]
```

```python
data12.problems.value_counts()
```

```python
data12.loc[data12.problems.isna()] = data12[data12.problems.isna()].fillna('None')
```

```python
data12.isna().sum()
```

```python
data12_Alive = data12[data12.status=='Alive']

data12_DeadORstump = data12[data12.status!='Alive']
```

## for 'Alive' trees

```python
stateAlive = data12_Alive.groupby('spc_latin').tree_dbh.describe().reset_index()[['spc_latin','25%','75%']]
stateAlive
```

```python
data12_Alive
```

```python
data12.status.value_counts()
```

```python
data12_Alive = data12_Alive.merge(stateAlive, on='spc_latin', how='left')
data12_Alive
```

```python
mask6 = data12_Alive.tree_dbh > data12_Alive['75%']
data12_Alive.loc[mask6, 'tree_dbh'] = data12_Alive['75%']

mask61 = data12_Alive.tree_dbh < data12_Alive['25%']
data12_Alive.loc[mask61, 'tree_dbh'] = data12_Alive['25%']
```
